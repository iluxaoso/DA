```python
import math
import numpy as np
import scipy.linalg as spl
from matplotlib import pylab as plt
```

```python
def f(x):
    return np.sin(x / 5.0) * (np.exp(x / 10.0)) + 5 * (np.exp(-x / 2.0))
```

```python
f(15)
```

```python
x1 = np.array([1,15])
A1 = lambda x: np.array([x ** i for i in range(2)]).T
A1 = A1(x1)
b1 = np.matrix([f(x1)]).T
#print(A1)
#print(b1)
w1 = spl.solve(A1,b1)
w1_0 = w1[0][0]
w1_1 = w1[1][0]
y1 = w1_0 + w1_1 * x1
```

```python
x = np.arange(1, 15.1, 0.1)
#print(x)
```

```python
plt.plot(x, f(x), x1, f(x1))
```

```python
x2 = np.array([1, 8, 15])
A2 = lambda x: np.array([x ** i for i in range(3)]).T
A2 = A2(x2)
b2 = np.matrix([f(x2)]).T
#print(A2)
#print(b2)
w2 = spl.solve(A2,b2)
w2_0 = w2[0][0]
w2_1 = w2[1][0]
w2_2 = w2[2][0]
y2 = w2_0 + w2_1 * x2 + w2_2 * (x2 ** 2)
```

```python
plt.plot(x, f(x), x2, f(x2))
```

```python
x3 = np.array([1, 4, 10, 15])
A3 = lambda x: np.array([x ** i for i in range(4)]).T
A3 = A3(x3)
b3 = np.matrix([f(x3)]).T
print(A3)
print(b3)
w3 = spl.solve(A3,b3)
print(w3)
w3[0] = w3[0][0]
w3[1] = w3[1][0]
w3[2] = w3[2][0]
w3[3] = w3[3][0]
y3 = w3[0] + w3[1] * x3 + w3[2] * (x3 ** 2) + w3[3] * (x3 ** 3)
```
