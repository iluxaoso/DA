Импортируем необходимые библиотеки.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

Построение гистограммы и графика плотности распределения для распределения $\chi2$

```python
x = np.linspace(0,30,1000)
k = 2
rv = sts.chi2(k)
#плотность распределения
pdf = rv.pdf(x)
plt.plot(x, pdf)
#гистограмма
sample = rv.rvs(1000)
plt.hist(sample, density=True)
plt.xlabel('Выборка')
plt.ylabel('Распределение (нормированное)')
```

Для n=[5,10,50,100,500] генерируются 1000 выборок объема n, находится выборочное среднее и строится гистограмма. Мат ожидание распределения $\chi2$=k, а дисперсия - 2k. Затем на графике с гистограммой строится функция плотности нормального распределения $N(k,\sqrt{\frac{2k}{n}})$

```python
n = 5
sample_5 = []
mean_5 = []
for j in range(1000):
    sample_5.append(rv.rvs(n))
    mean_5.append(sample_5[j].mean())
plt.hist(mean_5, bins=20, density=True)

mu = k
sigma = math.sqrt(2 * k / n)
norm_rv = sts.norm(loc=mu, scale=sigma)

min_val = min(mean_5)
max_val = max(mean_5)
x = np.linspace(min_val, max_val, 100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)
plt.xlabel('Выборочное среднее')
plt.ylabel('Распределение (нормированное)')
```
